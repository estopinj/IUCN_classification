# Loads .txt files

- Binary

```python
broad_p = "../logs/Selection_BROAD/broad_all_3var_activ/RF988_ntree10000_mtry100_2023-07-13 11:33:08.txt"

# *** PARAMS ***
stat = True
norm = "true"                     # {true, pred, all}, default=None
# **************

for stat in [True, False]:
    for norm in ["true", "pred", "all"]:

        with open(broad_p) as f:
            values = [l.strip() for l in f.readlines()]

        i_labs  = values.index("test.labels")
        i_preds = values.index("test.preds")
        N = 886

        labels = [int(v) for v in values[i_labs+1:i_labs+N+1]]
        preds  = [int(v) for v in values[i_preds+1:i_preds+N+1]]

        matrix = confusion_matrix(labels, preds, normalize=norm)
        print(matrix)

        labels = ['True N.','False P.','False N.','True P.']
        categories = ['Not Threat.', 'Threatened']
        make_confusion_matrix(matrix,
                              group_names=labels,
                              categories=categories,
                              count=False,
                              cmap='viridis', # "blues"
                              title='Binary classification',
                              sum_stats=stat,
                              vmin=0, vmax=0.86)

#         plt.savefig(os.path.join(fig_p, f"Binary_cfMatrix_stats{stat}_norm-{norm}.png") , bbox_inches="tight")
#         plt.savefig(os.path.join(fig_p, f"Binary_cfMatrix_stats{stat}_norm-{norm}.pdf") , bbox_inches="tight")
```

```python
broad_p = "../logs/Selection_DETAIL/detail_all_3var_activ/RF800_ntree10000_mtry100_2023-07-13 11:53:03.txt"

# *** PARAMS ***
stat = True
norm = "true"                     # {true, pred, all}, default=None
# **************

for stat in [True, False]:
    for norm in ["true", "pred", "all"]:

        with open(broad_p) as f:
            values = [l.strip() for l in f.readlines()]

        i_labs  = values.index("test.labels")
        i_preds = values.index("test.preds")
        N = 886

        labels = [int(v) for v in values[i_labs+1:i_labs+N+1]]
        preds  = [int(v) for v in values[i_preds+1:i_preds+N+1]]
        # preds

        matrix = confusion_matrix(labels, preds, normalize=norm)
        print(matrix)

        categories = ['LC','NT','VU','EN','CR']

        make_confusion_matrix(matrix,
                              categories=categories,
                              cmap='viridis',
                              count=False,
                              title='Status classification',
                              sum_stats=stat,
                              figsize=(10,7.5),
                              vmin=0, vmax=0.86)

#         plt.savefig(os.path.join(fig_p, f"Status_cfMatrix_stats{stat}_norm-{norm}.png") , bbox_inches="tight")
#         plt.savefig(os.path.join(fig_p, f"Status_cfMatrix_stats{stat}_norm-{norm}.pdf") , bbox_inches="tight")
```
