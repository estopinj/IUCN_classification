```python
import pandas as pd
import numpy as np
import os
os.environ['USE_PYGEOS'] = '0'
import geopandas as gpd

import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
```

- WGSRPD level 1

```python
lvl1_shp = "../data/wgsrpd/level1/level1.shp"
lvl1_gdf = gpd.read_file(lvl1_shp)

D1 = dict(zip(lvl1_gdf.LEVEL1_COD, lvl1_gdf.LEVEL1_NAM))
lvl1_gdf
```

- Occurrences reference

```python
p = "../occurrences/DeepOrchidSeriesElevHFP.csv"

df = pd.read_csv(p, sep=";")


df.rename(columns={"canonical_name":"species",
                   "decimallatitude":"lat",
                   "decimallongitude":"lon"
                  }, inplace=True)
df
```

- Groupby species

```python
sp = df.groupby("speciesKey").agg({
    "species":"first",
    "lat":"mean",
    "lon":"mean",
    "Elevation":"mean",
    "HFP09":"mean",
    "HFP93":"mean",
    "HFP_diff":"mean",
    "continent_code":lambda x: set(x)
})
sp["continents"] = sp.continent_code.apply(lambda S: set([ D1[s] for s in S]))
sp.drop(columns="continent_code", inplace=True)
sp
```

```python
spna = sp[np.isnan(sp.HFP_diff)]
sp   = sp[~np.isnan(sp.HFP_diff)]
print(len(spna), len(sp))
spna.head()
```

- Loads IUCN status

```python
status_p = "../data/status_export"
```

```python
names_ranks = ["Dispersal - present", "Dispersal - 2021/2040", "Dispersal - 2041/2060", "Dispersal - 2061/2080", "Dispersal - 2081/2100"]
ranks = ["rank_3var_activ", "r21.40_3var_activ", "r41.60_3var_activ", "r61.80_3var_activ", "r81.100_3var_activ"]
names_zeros = ["No disp. - present", "No disp. - 2021/2040","No disp. - 2041/2060","No disp. - 2061/2080","No disp. - 2081/2100"]
zeros = ["ZDpresent_3var_activ", "ZD21.40_3var_activ", "ZD41.60_3var_activ", "ZD61.80_3var_activ","ZD81.100_3var_activ"]
Ltp   = ['Present', '2021/2040', '2041/2060', '2061/2080', '2081/2100']

Ldf = []

for lvl in ["broad", "detail"]:
    for iucn_bool in ["TRUE", "FALSE"]:
        for scen,scen_name in zip(zeros+ranks, names_zeros+names_ranks):
            # name
            name = "_".join(["", lvl, scen, "iucnOnly", iucn_bool, ".csv"])
            df = pd.read_csv(os.path.join(status_p,name))
            df["level"]       = lvl
            df["IUCNonly"]   = iucn_bool
            df["Dispersal"]        = scen_name.split(' - ')[0]
            df["time_period"] = scen_name.split(' - ')[1]
            Ldf.append(df)
            
stat = pd.concat(Ldf)
stat.replace("FALSE", False, inplace=True)
stat.replace("TRUE", True, inplace=True)
stat.replace("present", "Present", inplace=True)
stat.replace("Dispersal", True, inplace=True)
stat.replace("No disp.", False, inplace=True)
stat.rename(columns={"status":"Status"}, inplace=True)
stat = stat.convert_dtypes()
stat
```

- Combines species dataframe with status

```python
dfp = sp.merge(stat, on="species")
dfp
```
